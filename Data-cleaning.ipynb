```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_2015 = pd.read_csv('2015.csv')
df_2016 = pd.read_csv('2016.csv')
df_2015.head(5)
```

```python
df_2016.head(5)
```

```python
#df_2015.insert(11, 'year', np.nan)
df_2016.insert(11, 'year', np.nan)
```

```python
df_2015.head(5)
df_2016.head(5)
```

```python
df_2015['year'].replace(np.nan, 2015, inplace=True)
df_2016['year'].replace(np.nan, 2016, inplace=True)
```

```python
df_2015.head(5)
```

```python
df = pd.concat([df_2015,df_2016])
df.sort_values(by='CT', ascending=True).head(5)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
filt = df['BPAD'].isna()
i = df.loc[filt].index
```

```python
df.drop(i, inplace=True)
```

```python
df.isna().sum()
df.shape
```

```python
data_dictionary = pd.read_excel('Data Dictionary.xlsx')
data_dictionary
```

```python
data_dictionary.drop(columns={'Unnamed: 1', 'Unnamed: 2'}, inplace=True)
data_dictionary
```

```python
data_dictionary.rename(columns={'Data Dictionary': 'Column', 'Unnamed: 3': 'Description of the column'}, inplace=True)
```

```python
data_dictionary.drop([0,1], inplace=True)
```

```python
new_row = {'Column': 'Year', 'Description of the column':'Year in which the Survey was done'}
new_row = pd.DataFrame(new_row, index=[1])
data_dictionary = pd.concat([data_dictionary, new_row], ignore_index=True)
data_dictionary.drop_duplicates(inplace=True)
data_dictionary.replace(np.nan, 'Blood Pressure Medication', inplace=True)
data_dictionary
```

```python
data_dictionary.to_excel('Cleaned Data Dictionary.xlsx', sheet_name='Data Dictionary', index=False)
```

```python
df_2015_cleaned = df[df['year'] == 2015]
df_2016_cleaned = df[df['year'] == 2016]

df_2015_cleaned.to_csv('2015-cleaned-dataset.csv', index=False)
df_2016_cleaned.to_csv('2016-cleaned-dataset.csv', index=False)
```

```python
df.to_csv('combined-dataset.csv', index=False)
df_2015_cleaned.columns
```

```python
df_2015_cleaned.rename(columns=
                       {
                           'CT' : 'CT_2015',
                           'BPAD':'BPAD_2015',
                           'BPAN':'BPAN_2015',
                           'BPAN2':'BPAN2_2015',
                           'BWAD': 'BWAD_2015',
                           'BWAN': 'BWAN_2015',
                           'BWAN2': 'BWAN2_2015',
                           'BMAD': 'BMAD_2015',
                           'BMAN': 'BMAN_2015',
                           'BMAN2': 'BMAN2_2015',
                           'year' : 'year_2015'
                       }, inplace=True)
df_2016_cleaned.rename(columns=
                       {
                           'CT' : 'CT_2016',
                           'BPAD':'BPAD_2016',
                           'BPAN':'BPAN_2016',
                           'BPAN2':'BPAN2_2016',
                           'BWAD': 'BWAD_2016',
                           'BWAN': 'BWAN_2016',
                           'BWAN2': 'BWAN2_2016',
                           'BMAD': 'BMAD_2016',
                           'BMAN': 'BMAN_2016',
                           'BMAN2': 'BMAN2_2016',
                           'year' : 'year_2016'
                       }, inplace=True)
```

```python
df_merge = pd.merge(df_2015_cleaned, df_2016_cleaned, on='_id', how='left')
```

```python
df_merge.columns
df_merge.shape
df_merge.to_csv('Merged-dataset-2015-and-2016.csv', index=False)
```
